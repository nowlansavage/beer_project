```python
# Pandas
import pandas as pd
    
# SQL Alchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
```

```python
# create engine for untappd

untappd_database_path = 'Sqlite_files/untappd.sqlite'
engine = create_engine(f"sqlite:///{untappd_database_path}")
conn = engine.connect()
```

```python
untappd_data = pd.read_sql('SELECT * FROM untappd', conn)
untappd_data.head()
```

```python
# create engine for beer advocate 
advocate_database_path = 'Sqlite_files/beer_advocate_2017.sqlite'
engine = create_engine(f"sqlite:///{advocate_database_path}")
conn = engine.connect()
```

```python
inspector = inspect(engine)
```

```python
inspector.get_table_names()

advocate_data = pd.read_sql('SELECT * FROM beer_advocate_2017_ratings', conn)
advocate_data.head()
```

```python
advocate_data.columns=['index','brewery','beer','brewery_beer','style','abv','rating']
advocate_data.head()

```

```python
bad_cell = advocate_data[advocate_data['index']==277771].index
bad_cell
advocate_data.drop(bad_cell, inplace=True)
advocate_data.head()
```
