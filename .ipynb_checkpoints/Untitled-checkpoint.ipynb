```python
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
breweries = {}

url = "https://www.brewersassociation.org/directories/breweries/"
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, "html.parser")
```

```python
breweries["name"] = soup.find("h3", class_="with-mini-hr")
# breweries["website"] = soup.find("span", class_="result-price").get_text()
# breweries["address"] = soup.find("span", class_="result-hood").get_text()

# return breweries
breweries
```
