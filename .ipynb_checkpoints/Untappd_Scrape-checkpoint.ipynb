```python
from splinter import Browser
from bs4 import BeautifulSoup
import csv
import pandas as pd
import csv_to_sqlite
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.tasteatlas.com/most-popular-beers-in-the-world'
browser.visit(url)
```

```python
styleslist=[]

soup = BeautifulSoup(browser.html,'html.parser')
beer_styles=soup.find_all('div', class_='middle')
    
for beer_style in beer_styles:
    style = beer_style.find('h2', class_='h1 h1--bold').get_text("|", strip=True).replace(' ','+')
    styleslist.append(style)
```

```python
print(styleslist)
```

```python
# TRIED TO AUTOLOGIN AT UNTAPPD, BUT HAS RECAPTCHA AND CAN'T.  WITHOUT LOGIN, YOU ONLY GET 5 OF EACH BEER STYLE

# # takes the email address for the untappd account
# user_name = raw_input("spearjay ")

# # takes the oassword for the user account needed
# user_pass = raw_input("Generic2020! ")

# # loads the chrome broswer
# browser= Browser('chrome')

# #selects untappd as the website to load in the browser
# browser.visit('https://untappd.com/')

# # fills the user field in the untappd login section 
# browser.fill('Username', user_email)
# browser.fill('Password', user_pass)

# <div class="recaptcha-checkbox-border" role="presentation" style=""></div>
```

```python
top_five_style_beers_untappd = {}
top_five_style_beers_untappd_df = pd.DataFrame({'beer': [], 'brewery': [], 'style': [],'link': [], 'abv': [], 'rating': []})

for style in styleslist:
    url = (f'https://untappd.com/search?q={style}')
    print(url)
    browser.visit(url)

    soup = BeautifulSoup(browser.html,'html.parser')
    details = soup.find_all('div', class_='beer-item')

    for detail in details:
        beer = detail.find('p', class_='name').get_text()
        brewery = detail.find_all('p',class_='style')[0].get_text()
#         style = detail.find_all('p',class_='style')[1].get_text()
        link = detail.find('p').find('a')['href']
        abv = detail.find('p', class_='abv').get_text("|", strip=True)
        rating = detail.find('span', class_='num').get_text("|", strip=True).replace('(','').replace(')','')
        top_five_style_beers_untappd_df = top_five_style_beers_untappd_df.append({'beer': beer, 'brewery': brewery, 'style': style, 'link': link, 'abv': abv, 'rating': rating}, ignore_index=True)

        top_five_style_beers_untappd_df.append(top_five_style_beers_untappd_df)
```
