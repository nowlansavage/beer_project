```python
url = 'https://untappd.com/beer/top_rated'
browser.visit(url)
```

```python
top_beers_untappd ={}
top_beers_untappd_df = pd.DataFrame({'beer':[],'brewery':[],'style':[],'link':[],'abv':[],'rating':[]})
# url = (f'https://untappd.com/search?q={style}')
# browser.visit(url)

soup = BeautifulSoup(browser.html,'html.parser')
details = soup.find_all('div', class_='beer-item')

for detail in details:
    beer = detail.find('p', class_='name').get_text()
    brewery = detail.find_all('p',class_='style')[0].get_text()
    style = detail.find_all('p',class_='style')[1].get_text()
    link = detail.find('p').find('a')['href']
    abv = detail.find('p', class_='abv').get_text("|", strip=True)
    rating = detail.find('span', class_='num').get_text("|", strip=True).replace('(','').replace(')','')
    top_beers_untappd_df = top_beers_untappd_df.append({'beer': beer, 'brewery': brewery, 'style': style, 'link': link, 'abv': abv, 'rating': rating}, ignore_index=True)

#     top_beers_untappd_df.append(top_beers_untappd_df)
```

```python
top_beers_untappd_df.to_csv('untappd.csv')
```
