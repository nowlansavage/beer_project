```python
from splinter import Browser
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import requests
import pandas as pd

from sqlalchemy import create_engine
engine = create_engine('sqlite:///beer_project.sqlite', echo=False)
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
# browser = Browser('chrome', **executable_path, headless=False)
driver = webdriver.Chrome('chromedriver.exe')
```

```python

url = "https://www.brewersassociation.org/directories/breweries/"
# browser.visit(url)
driver.get(url)

# Used code from https://stackoverflow.com/questions/32391303/how-to-scroll-to-the-end-of-the-page-using-selenium-in-python
lenOfPage = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
match=False
while(match==False):
        lastCount = lenOfPage
        time.sleep(3)
        lenOfPage = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
        if lastCount==lenOfPage:
            match=True
```

```python
html = driver.page_source
# html = driver.html
soup = BeautifulSoup(html, "html.parser")
soup
```

```python
results = soup.find_all("div", class_="company-listing")
name_list = []
url_list = []
address_list = []
city_list = []
state_list = []
for result in results:
    names = result.find("h3", class_="with-mini-hr")
    websites = result.find("a", itemprop="image")
    addresses = result.find("p", itemprop="streetAddress")
    cities = result.find("span", itemprop="addressLocality")
    states = result.find("span", itemprop="addressRegion")
    if names:
        name_list.append(names.text)     
    else:
        name_list.append("")
    
    if websites:
        url_list.append(websites.text)
    else:
        url_list.append("")
   
    if addresses:
        address_list.append(addresses.text)
    else:
        address_list.append("")
        
    if cities:
        city_list.append(cities.text)
    else:
        city_list.append("")
        
    if states:
        state_list.append(state.text)
    else:
        state_list.append("")
#     if website"
        
#     brewery = {
#         'name' : names,
#         'website' : websites,
#         'address' : addresses,
#         'city' : cities,
#         'state' : states
#     }
#     collection.insert_one(brewery)
brewery = {
    'name' : name_list,
    'website' : url_list,
    'address' : address_list,
    'city' : city_list,
    'state' : state_list
}
brewery
```

```python
breweries_df = pd.DataFrame.from_dict(brewery)

breweries_df.to_sql('breweries', con=engine)
```
